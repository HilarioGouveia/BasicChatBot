```python
# Importa biblioteca
from google.colab import ai # LLMs google/gemini-2.5-flash

# Atribui o valor do input a variável "pergunta"
pergunta = input('Pergunte para IA: ')

# Atribui o valor do texto gerado pela IA a variável "resposta"
resposta = ai.generate_text(pergunta)

# Imprime a resposta

print(f'\n''Resposta da IA: {resposta}', '\n')
```
